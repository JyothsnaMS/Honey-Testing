```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Glucose', 'Fructose', 'Sucrose', 'Moisture', 'Reducing_Sugar', 'FR_Ratio']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_range, sse)
```
